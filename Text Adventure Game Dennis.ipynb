<h2>Introduction</h2>
<br>
<p>The following text adventure game is created based on the TV series Merlin, season one episode 4. The game is extremely loose that the only common things are the characters names, a few scenes and the actual mission. </p>
<p>Merlin is almost dying after drinking Arthur's goblet which had been poisoned by Nimueh, a sorcerer who posed as a Mercian handmaiden named Cara so that she could poison prince Arthur. Gaius, a witch doctor of the palace, discovers a poisonous flower petal stuck inside the chalice and explains to Arthur that an antidote can only be made from the leaves from the same flower, the Mortaeus which is found in the caves beyond the Forests of Balor. Gaius also adds that going to the forest would be dangerous and no one might make it back. Arthur insists on going and Gaius leaves him with one advice, that he should not trust any wild animal in the forest since they do not exist and if he sees one it might not be real. The mission of the game is to help Arthur to go to the forest and get the flower before Merlin dies.</p>
<br>
<h2>Game Structure</h2>
<br>
<p>The game consists of three stages, a win and a fail global define functions. No bugs and errors.</p>
<p><b>First Stage</b> - The player needs to help Arthur figure out which path to follow since there are two paths heading into the forest.</p>
<p><b>Second Stage</b> - Inside the forest, Arthur meets with Nimueh disguised as a maiden claiming to have run away from her abusive master. He needs to make a decision whether to trust the maiden, fight a Cockatrice standing on his way or look for another alternative. Remember to consider time in your decision.</p>
<p><b>Third Stage</b> - The player needs to help Arthur make a wise decision since Nimueh has casted a spell making it difficult for Arthur to secure the flower.</p>

```python
#importing packages
import random
import time
import sys
#defining a win function
def win():
    print(f"""
{"*"*80}

                              YOU WIN!

                  Thank you For Helping Arthur!!!!!!!
                                                         _ 
              /\                                        | |
             /  \__      _____  ___  ___  _ __ ___   ___| |
            / /\ \ \ /\ / / _ \/ __|/ _ \| '_ ` _ \ / _ \ |
           / ____ \ V  V /  __/\__ \ (_) | | | | | |  __/_|
          /_/    \_\_/\_/ \___||___/\___/|_| |_| |_|\___(_)
                                                    
    
{"*"*80}""")
    #controlling the input attempts for the replay action
    invalid_entries = 2 #setting the number of entries needed
    #using a while loop to control the user entries
    while invalid_entries > 0:
        replaygame = input(prompt = "Would like to start again? (Yes or No)\n")
        if replaygame.casefold() == "yes" or "y" in replaygame.casefold():
            stage1()
            break
        elif replaygame.casefold() == "no" or "n" in replaygame.casefold():
            print(f"""
{"*"*80}

                              EXIT GAME!
                                                        
{"*"*80}""")
            break
        else:
            invalid_entries -= 1
            if invalid_entries > 0:
                print(f"I didn't get that! Please you have {invalid_entries} more chance to accept or exit.")
            elif invalid_entries == 0:
                print(f"You have {invalid_entries} chance left. Bye Bye!")
                #printing exit game if the invalid entries are over 
                print(f"""
{"*"*80}

                              EXIT GAME!
                                                        
{"*"*80}""")
                      
            else:
                print("Error occured!")     

    
    
#defining a fail function
def fail():
    print(f"""
{"*"*80}

                              YOU LOST!

                            GAME OVER!!!!!!!
               _____                                            
              / ____|                                           
              | |  __  __ _ _ __ ___   ___    _____   _____ _ __ 
              | | |_ |/ _` | '_ ` _ \ / _ \  / _ \ \ / / _ \ '__|
              | |__| | (_| | | | | | |  __/ | (_) \ V /  __/ |   
               \_____|\__,_|_| |_| |_|\___|  \___/ \_/ \___|_|   
                                                    
    
{"*"*80}""")
    #controlling the invalid entries in the repeat game function
    invalid_entries = 2 #setting the number of attempts for invalid input
    while invalid_entries > 0:
        repeatgame = input(prompt = "Would like to start again? (Yes or No)\n")
        if repeatgame.casefold() == "yes" or "y" in repeatgame.casefold():
            stage1()
            #ending the loop when the condition is met
            break
        elif repeatgame.casefold() == "no" or "n" in repeatgame.casefold():
            print(f"""
{"*"*80}

                              EXIT GAME!
                                                        
{"*"*80}""")#ending the loop when the condition is met
            break
        else:
            invalid_entries -= 1 #subtrating every invalid attempts made
            if invalid_entries > 0:
                print(f"I didn't get that! Please you have {invalid_entries} more chance to accept or exit.")
            elif invalid_entries == 0:
                print(f"You have {invalid_entries} chance left. Bye Bye!")
                print(f"""
{"*"*80}

                              EXIT GAME!
                                                        
{"*"*80}""")
                      
            else:
                print("Error occured!")     
        
    
    

    
    
#defining the start of the game    
def stage1():    
    def choosepath():
        paths_count = 1 #setting the number of trial for stage one
        #asking the player to choose the path to follow
        choosepath1 = input(prompt = f"""
    Which path will you choose? 
    Remember you only have 1 trial.
    1.) Path 1
    2.) Path 2\n""")
        #adding a while loop for the path options and setting the trial function
        while paths_count > 0:
            paths_count -= 1
            if "2" in choosepath1 or choosepath1.casefold() == "two" or choosepath1.casefold() == "path 2":
                print("""
    Great work! You are right! Gaius talked about the forest being dangerous.""")
                
                print("\n")
                input("<Press Enter To Continue>\n")
                stage2()
                break
                
            elif "1" in choosepath1 or choosepath1.casefold() == "one" or choosepath1.casefold() == "path 1":
                print(f"""
    That's incorrect and you have {paths_count} trials remaining.
    You didn't follow Gaius instructions. You fail!""")
                fail()
            else:
                print("Sorry! I didn't get that right! Invalid Option!")
                fail()               
        
        
    
    #asking the player if he or she would like to know more about the paths with a condition state
    #printing the welcome statement of the game
    print(f"""
{"*"*80}

           Welcome to King Arthur and the Quest of the Mortaeus Flower!!! 
                                STAGE 1!
    
{"*"*80}""")
    print("\n")
    input("<Press Enter or To Continue>\n")
    print("""
    With Merlin dying, after drinking the contents of the goblet which was 
    poisoned by sorcerer Nimueh and was meant for Arthur, Gaius discovers a poisonous
    flower petal stuck inside the chalice and explains to Arthur that an antidote
    can only be made from the leaves from the same flower, the Mortaeus flower,
    which may only be found in the caves beyond the Forests of Balor. 
    He also explains that it would be a dangerous journey although he will encounter wild things which are not 
    real but seems real.""")
    print("\n")
    input("<Press Enter or To Continue>\n")
    #printing out what is required in stage one
    print("""
    On his way to find the caves beyond the Forests of Balor, Arthur comes 
    across two roads heading into the forest. Help Arthur figure out which 
    road he should use.""")
    time.sleep(1) #giving the user time to read the requirements with a time package
    print("\n")#introducing a new line
    #defining the paths
    pathdescription = """
    Path One: The path contains fresh growing trees with footprints of people
    showing that there has been movement in and outside the forest. If you look
    deep into the road it seems there is a way through the forest.
    
    Path Two: The path looks dark, bushy and doesn't seems like there is any evidence 
    of a living thing. It is foggy and very quiet. \n"""
    #asking the player if he or she would like to know more about the paths with a condition state
    incorrect_trial = 3
    while incorrect_trial > 0:
        description_input = input(prompt = "Would you like to know more about the paths? (Yes or No)")
        if description_input.casefold() == "Yes" or "y" in description_input.casefold():
            description = pathdescription
            #a for loop for to print each letter in the path description after every 0.03 seconds
            for char in description:
                sys.stdout.write(char)
                sys.stdout.flush()
                time.sleep(0.03)
            choosepath()#calling the local defined function choosepath
            break #stoping the while loop when condition is met
        elif description_input.casefold() == "No" or "n" in description_input.casefold():
            print("Oh well! It seems you are in for a gamble with the paths! Goodluck!")
            #calling the local defined function choosepath
            choosepath()
             #stoping the while loop when condition is met
            break
        else:
            #controlling the incorrect trial everytime anything is entered apart from yes or no
            incorrect_trial -= 1
            if incorrect_trial > 0:
                print(f"\nI didn't get you! You have {incorrect_trial} trials remaining! Please Input Yes or No\n")
            elif incorrect_trial == 0:
                print(f"\nI didn't get you! You have {incorrect_trial} trials remaining! Follow Instructions\n")
                fail()
            else:
                print("Error!")
#defining stage two of the game
def stage2():
    #defining the nested defined function user_input
    def user_inputstage2():
        #list of what is in the choices
        choice_lst =["ignore Nimueh and find the caves on his own",
                    "take a chance on Nimueh's offer",
                    "fight the Cockatrice and proceed"]
        #setting the number of guesses or incorrect answers to be made
        guesses = 3
        #a while loop to control the invalid entries
        while guesses > 0:
            choice = input("""\n
    Please help arthur make the right choice (1-3)?\n>
    1. Ignore Nimueh and find the caves on his own
    2. Take a chance on Nimueh's offer
    3. Fight the Cockatrice and proceed\n""")
            #controling the inputs to an integer
    
            try:
               # converting choice for numerical comparison
                choice = int(choice)
    
            except:
            #if the inputs are not an integer then we reduce it after every invalid answer
                guesses -= 1
                #nested condition statement to control the strings entered
                if guesses > 0:
                    print(f"""
    You have {guesses} guesses remaining.                 
    I need a number in digits.\n""")
                    continue
                elif guesses == 0:
                    print(f"""
    You have {guesses} guesses remaining.                
    I need a number in digits. Press enter to try again.\n""")
                    fail()
                else:
                    print("Error!")
    
        # conditional statement based on user input
            if choice == 2:
                print(f"""
    That's right! Let's {choice_lst[1]} and find the 
    cave in time to save Merlin's life.""")
                print("Great Work!")
                print("\n")
                input("<Press Enter to Continue>")
                #directing the player to stage 3 if condition is met
                stage3()
                #stoping the while loop when condition is met
                break

            elif choice == 1 or choice == 3:
                #a control condition for the two incorrect choices
                guesses -= 1
                if guesses > 0:
                    if choice == 1:
                        print(f"\nThat's incorrect. You have {guesses} guesses remaining.")
                        print(f"""\nTo {choice_lst[0]} will
                        take sometime and Merlin is dying.""")
                    else:
                        print(f"\nThat's incorrect. You have {guesses} guesses remaining.")
                        print(f"""
    \nTo {choice_lst[2]}, you
    are not sure if you will kill it and save Merlin.""")
                elif guesses == 0:
                    print(f"\nYou are out of guesses. The choice was: 2. {choice_lst[1]}.")
                    fail()
                else:
                    print("Error!")

        # setting a different error message for inner and outer loops
            else:
                print("\nSomething went wrong.")
    #welcome message for stage 2            
    print(f"""
{"*"*80}
                     STAGE 2!
           I'm glad you could make it to stage 2.                   
    
{"*"*80}""")
    print("\n")
    print("Another mystery awaits!")
    #Printing a description for stage 2
    print("""
    Inside the forest, Arthur meets Nimueh disguised as a maiden 
    claiming to have a run away from her abusive master. She tells Arthur
    that she knows where the caves are and can lead him there easily.
    At a distance Arthur also sees a Cockatrice standing on his way and he
    must fight it if he want to continue with the journey on the same direction.""")            
    user_inputstage2()
#defining stage three of the game
def stage3():
    #defining the nested defined function user_input
    def user_inputstage3():
        #list of what is to be decided
        decision_lst =["walk towards the flower and grab it",
                    "fight the two lions and secure the flower",
                    "give up and let Merlin die"]
        #setting the number of guesses or incorrect answers to be made
        guesses = 3
         #a while loop to control the invalid entries
        while guesses > 0:
            decision = input("""\n
    What should Arthur do? (1-3)?\n>
    1. Walk towards the flower and grab it.
    2. Fight the two lions and secure the flower.
    3. Give up and let Merlin die\n""")
            #controling the inputs to an integer
    
            try:
               # converting choice for numerical comparison
                decision = int(decision)
    
            except:
            #if the inputs are not an integer then we reduce it after every invalid answer
                guesses -= 1
                #nested condition statement to control the strings entered
                if guesses > 0:
                    print(f"""
    You have {guesses} guesses remaining.                 
    I need a number in digits.\n""")
                    continue
                elif guesses == 0:
                    print(f"""
    You have {guesses} guesses remaining.                
    I need a number in digits. Press enter to try again.\n""")
                    fail()
                else:
                    print("Error!")
    
        # conditional statement based on user input
            if decision == 1:
                print(f"""That's right! Let's {decision_lst[0]}. Gaius talked 
     about the cave being scary with wild animals that are not real due to 
     sorcery.\n""")
                print("You've helped Arthur to secure the flower!")
                print("\n")
                input("<Press Enter to Continue>")
                #directing the player to win function if condition is met
                win()
                break

            elif decision == 2 or decision == 3:
                #a control condition for the two incorrect choices
                guesses -= 1
                if guesses > 0:
                    if decision == 2:
                        print(f"\nThat's incorrect. You have {guesses} guesses remaining.")
                        print(f"""\nTo {decision_lst[1]} will
      waste time since they are not real and is just a form of distraction from Nimueh.""")
                    else:
                        print(f"\nThat's incorrect. You have {guesses} guesses remaining.")
                        print(f"""To {decision_lst[2]} will jeopardize your mission.""")
                elif guesses == 0:
                    print(f"\nYou are out of guesses. The choice was: 2. {choice_lst[0]}.")
                    fail()
                else:
                    print("Error!")

        # setting a different error message for inner and outer loops
            else:
                print("\nSomething went wrong.")
    #welcome message for stage 3            
    print(f"""
{"<"*40}{">"*40}
                            STAGE 3!
           Awesome! You've made it to the final Stage.                   
    
{"<"*40}{">"*40}""")

    print("\n")
    print("You are almost there!")
    print("\n")
    #Printing a description for stage 3
    print("""
    Arthur Gets to the cave with Nimueh and sees the flowers.
    Nimueh casts a spell bringing two very big wild animals near the flower.
    Help Arthur to secure the flower.""")
    print("\n")
    #calling the local defined function user_inut
    user_inputstage3()
#calling the start of the game as global function     
stage1() 

```
